```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from google.colab import auth
import gspread
from google.auth import default
```

```python
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
worksheet = gc.open('training data').sheet1
```

```python
data = worksheet.get_all_values()
dataset1 = pd.DataFrame(data[1:], columns = data[0])
dataset1 = dataset1.astype({'Inputs':'float'})
dataset1 = dataset1.astype({'Outputs':'float'})
dataset1.head()
```

```python
X = dataset1[['Inputs']].values
y = dataset1[['Outputs']].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 33)
Scaler = MinMaxScaler()
Scaler.fit(X_train)
X_train1 = Scaler.transform(X_train)
ai_brain = Sequential([
    Dense(16,activation = 'relu'),
    Dense(32,activation = 'relu'),
    Dense(1)
])

ai_brain.compile(optimizer = 'rmsprop', loss = 'mse')
ai_brain.fit(X_train1, y_train,epochs = 1000)
```

```python
X = dataset1[['Inputs']].values
y = dataset1[['Outputs']].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 33)
Scaler = MinMaxScaler()
Scaler.fit(X_train)
X_train1 = Scaler.transform(X_train)
ai_brain = Sequential([
    Dense(16,activation = 'relu'),
    Dense(32,activation = 'relu'),
    Dense(1)
])

ai_brain.compile(optimizer = 'rmsprop', loss = 'mse')
ai_brain.fit(X_train1, y_train,epochs = 1000)
```

```python
loss_df = pd.DataFrame(ai_brain.history.history)
loss_df.plot()
```

```python
X_test = Scaler.transform(X_test)
ai_brain.evaluate(X_test,y_test)
```
